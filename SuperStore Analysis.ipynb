# What is the sales trend over time (monthly, yearly)?

```python
df.info()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'],errors='coerce')
```

#### Sales over time

```python
monthly_sales=df.groupby(['Order Date'],as_index=False).sum(numeric_only=True)
```

```python
monthly_sales=monthly_sales.set_index('Order Date')
```

```python
monthly_sales = monthly_sales.resample('M').sum()
```

```python
monthly_sales
```

```python
plt.figure(figsize=(20,10))
plt.plot(monthly_sales['Sales'])
plt.title('Monthly Sales Trend')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.show()
```

```python
# Years Sales
year_sales=monthly_sales.resample('Y').sum()
```

```python
plt.figure(figsize=(20,10))
plt.plot(year_sales['Sales'])
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.title('Yearly Sales Trend')
plt.show()
```

#### Profit Over Time

```python
monthly_sales=df.groupby(['Order Date'],as_index=False).sum(numeric_only=True)
monthly_sales=monthly_sales.set_index('Order Date')
monthly_sales=monthly_sales.resample('M').sum()
plt.figure(figsize=(25,10))
plt.plot(monthly_sales['Profit'])
plt.xlabel('Order Date')
plt.ylabel('Profit')
plt.title('Monthly Profit Trend')
plt.show()
```

```python
yearly_profit=monthly_sales.resample('Y').sum()
plt.figure(figsize=(25,10))
plt.plot(yearly_profit['Profit'])
plt.xlabel('Order Date')
plt.ylabel('Profit')
plt.title('Yearly Profit Trend')
plt.show()
```

# Which region & place generates the most sales?

```python
df.columns
```

```python
df_place=df[['Country', 'City', 'State','Region']]
df_place.head()
```

```python
for j in df_place.columns:
    print(j,":",df_place[j].nunique())
```

```python
df_place=df[['Country', 'City', 'State','Region','Sales','Profit']]
df_place.head()
```

```python
region_sales=df_place.groupby(['Region'],as_index=False).sum(numeric_only=True)
region_sales.sort_values(by='Sales',ascending=False,inplace=True)
plt.figure(figsize=(15,5))
plt.bar(region_sales['Region'],region_sales['Sales'],align='center')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.title('Sales generated by Region')
plt.show()
```

```python
region_profit=df_place.groupby(['Region'],as_index=False).sum(numeric_only=True)
region_profit.sort_values(by='Profit',ascending=False,inplace=True)
plt.figure(figsize=(15,5))
plt.bar(region_profit['Region'],region_profit['Profit'],align='center')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.title('Profit generated by Region')
plt.show()
```

```python
state_sales=df_place.groupby(['State'],as_index=False).sum(numeric_only=True)
state_sales.sort_values(by='Sales',ascending=False,inplace=True)
plt.figure(figsize=(25,10))
plt.bar(state_sales['State'],state_sales['Sales'],align='center')
plt.xlabel('State')
plt.ylabel('Sales')
plt.title('Sales generated by State')
plt.xticks(rotation=90)
plt.show()
```

```python
city_sales=df_place.groupby(['City'],as_index=False).sum(numeric_only=True)
city_sales.sort_values(by='Sales',ascending=False,inplace=True)
plt.figure(figsize=(25,10))
plt.bar(city_sales['City'].head(),city_sales['Sales'].head(),align='center')
plt.xlabel('City')
plt.ylabel('Sales')
plt.title('Top 5 Cities By Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
city_profit=df_place.groupby(['City'],as_index=False).sum(numeric_only=True)
city_profit.sort_values(by='Profit',ascending=False,inplace=True)
plt.figure(figsize=(25,10))
plt.bar(city_profit['City'].head(),city_profit['Profit'].head(),align='center')
plt.xlabel('City')
plt.ylabel('Profit')
plt.title('Top 5 Cities By Profit')
plt.xticks(rotation=90)
plt.show()
```

# What is the impact of discounts on sales?

```python
discount_group=df.groupby(['Discount']).sum(numeric_only=True)[['Sales']]
ax=discount_group.plot(kind='bar')
ax.set_ylabel('Sales')
plt.show()
```

```python
plt.scatter(df['Discount'],df['Sales'])
plt.xlabel('Discount')
plt.ylabel('Sales')
plt.show()
```

```python
# verage profit for each category
avg_profit_by_category=df.groupby(['Category'],as_index=False).mean(numeric_only=True)[['Category','Profit']]
avg_profit_by_category
```

```python
df['Profit Margin']=df['Profit']/df['Sales']
```

```python
avg_profit_margin_by_category=df.groupby(['Category']).mean(numeric_only=True)[['Profit Margin']]
avg_profit_margin_by_category.plot(kind='bar')
plt.xlabel('Product Category')
plt.ylabel('Average Profit Margin')
plt.title("Average Profit Margin by Product Category")
plt.show()
```
